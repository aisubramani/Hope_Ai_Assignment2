```python
from openai import OpenAI


client = OpenAI(
  api_key="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"
)

messages = []
system_msg = input("What type of chatbot would you like to create?\n")
messages.append({"role": "system", "content": system_msg})

print("Your new assistant is ready! (press 'q' to quit)")

while True:
    user_msg = input()

    # EXIT CONDITION HERE
    if user_msg.lower().strip() == "q":
        print("Chatbot closed!")
        break

    messages.append({"role": "user", "content": user_msg})

    response = client.chat.completions.create(
        model="gpt-5-nano",
        messages=messages
    )

    reply = response.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})

    print("\nAssistant:", reply, "\n")
```
