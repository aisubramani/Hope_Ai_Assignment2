```python
import pandas as pd
import re

# Load the text file
with open("mabel.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

data = []

# Regex pattern for WhatsApp chat format
# 05/12/19, 1:42 pm - Mabel Infoziant: Hi this is Mabel we just spoke
#()	- Creates 1 column
#re.match()	- Applies regex
#match.groups() -Returns 4 values
#date, time, person, message =	Splits into 4 variables
#data.append()-Adds as one table row
pattern = r"(\d{2}/\d{2}/\d{2}),\s(\d{1,2}:\d{2}\s[ap]m)\s-\s(.*?):\s(.*)"

for line in lines:
    match = re.match(pattern, line)
    if match:
        date, time, person, message = match.groups()
        data.append([date, time, person, message])

# Create DataFrame
df = pd.DataFrame(data, columns=["Date", "Time", "Name", "Chat"])

# Show table
#print(df)

# Optional: Save as CSV
#df.to_csv("mabel_chat_table.csv", index=False)

print("\nTable created and saved successfully!")
df
```

```python
import pandas as pd

# Load the text file
with open("mabel.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

data = []

for line in lines:
    line = line.strip()
    # Example:
    # "05/12/19, 1:42 pm - Mabel Infoziant: Hi this is Mabel we just spoke"

    # Skip system messages automatically
    if " - " in line and ": " in line:

        # Split Date
        date_part, rest1 = line.split(", ", 1)
        # "05/12/19" , "1:42 pm - Mabel Infoziant: Hi this is Mabel we just spoke"

        # Split Time
        time_part, rest2 = rest1.split(" - ", 1)
        # "1:42 pm" , "Mabel Infoziant: Hi this is Mabel we just spoke"

        # Split Person & Chat Message
        name, chat = rest2.split(": ", 1)
        # "Mabel Infoziant" , "Hi this is Mabel we just spoke"

        data.append([date_part, time_part, name, chat])

# Create DataFrame
df = pd.DataFrame(data, columns=["Date", "Time", "Name", "Chat"])

# Save as CSV 
#df.to_csv("mabel_chat_table.csv", index=False)

print("\nTable created successfully using SIMPLE split()!\n")

# Display DataFrame (Jupyter/Colab)
df
```

```python
#Apply VADER Sentiment Analysis
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
```

```python
# Initialize VADER
sid = SentimentIntensityAnalyzer()

# Apply sentiment on Chat column
df["scores"] = df["Chat"].apply(lambda x: sid.polarity_scores(x))

# Split scores into separate columns
df["compound"] = df["scores"].apply(lambda x: x["compound"])
df["Negative"] = df["scores"].apply(lambda x: x["neg"])
df["Positive"] = df["scores"].apply(lambda x: x["pos"])
df["Neutral"]  = df["scores"].apply(lambda x: x["neu"])
df
```

```python
# Create Your Custom comp_score (Label)
def sentiment_label(score):
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df["comp_score"] = df["compound"].apply(sentiment_label)
df
```

```python
import pandas as pd
import nltk
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# -------------------------------
# DOWNLOAD VADER (Only first time)
# -------------------------------
nltk.download('vader_lexicon')

# -------------------------------
# LOAD WHATSAPP CHAT FILE
# -------------------------------
with open("mabel.txt", "r", encoding="utf-8") as file:
    lines = file.readlines()

data = []

for line in lines:
    line = line.strip()

    # Skip system messages
    if " - " in line and ": " in line:

        # Split Date
        date_part, rest1 = line.split(", ", 1)

        # Split Time
        time_part, rest2 = rest1.split(" - ", 1)

        # Split Name & Chat
        name, chat = rest2.split(": ", 1)

        data.append([date_part, time_part, name, chat])

# -------------------------------
# CREATE DATAFRAME
# -------------------------------
df = pd.DataFrame(data, columns=["Date", "Time", "Name", "Chat"])

# -------------------------------
# APPLY VADER SENTIMENT ANALYSIS
# -------------------------------
sid = SentimentIntensityAnalyzer()

df["scores"] = df["Chat"].apply(lambda x: sid.polarity_scores(x))
df["compound"] = df["scores"].apply(lambda x: x["compound"])
df["Negative"] = df["scores"].apply(lambda x: x["neg"])
df["Positive"] = df["scores"].apply(lambda x: x["pos"])
df["Neutral"]  = df["scores"].apply(lambda x: x["neu"])

# -------------------------------
# CREATE FINAL SENTIMENT LABEL
# -------------------------------
def sentiment_label(score):
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df["comp_score"] = df["compound"].apply(sentiment_label)

# -------------------------------
# OVERALL CHAT SENTIMENT
# -------------------------------
overall = df["compound"].mean()

if overall >= 0.05:
    print("‚úÖ Overall Chat Sentiment: POSITIVE")
elif overall <= -0.05:
    print("‚ùå Overall Chat Sentiment: NEGATIVE")
else:
    print("‚öñ Overall Chat Sentiment: NEUTRAL")

print("Overall Compound Score:", round(overall, 3))

# -------------------------------
# SENTIMENT BAR CHART (NO COLORS)
# -------------------------------
sentiment_counts = df["comp_score"].value_counts()

plt.figure()
sentiment_counts.plot(kind="bar")
plt.title("Overall Chat Sentiment Distribution")
plt.xlabel("Sentiment Type")
plt.ylabel("Number of Messages")
plt.tight_layout()
plt.show()

# -------------------------------
# SAVE FINAL RESULT
# -------------------------------
df.to_csv("mabel_chat_sentiment_analysis.csv", index=False)

# -------------------------------
# DISPLAY FINAL DATA
# -------------------------------
print("\n‚úÖ Sentiment Analysis Completed Successfully!\n")
df
```
