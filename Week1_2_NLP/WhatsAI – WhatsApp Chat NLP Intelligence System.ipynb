# Android WhatsApp Line Parsing Logic

```python
import re
import pandas as pd
from datetime import datetime


# =======================================================
# REGEX PATTERN FOR ANDROID WHATSAPP MESSAGES
# =======================================================

msg_pattern = re.compile(
    r'^(\d{1,2}/\d{1,2}/\d{2,4}),\s(\d{1,2}:\d{2})\s?(am|pm|AM|PM)?\s-\s(.*)$'
)


# =======================================================
# PARSER FUNCTION
# =======================================================

def parse_whatsapp_chat(lines):
    messages = []

    current_msg = {
        "datetime": None,
        "date": None,
        "time": None,
        "sender": None,
        "message": "",
        "is_system": False
    }

    for raw_line in lines:
        line = raw_line.strip()

        match = msg_pattern.match(line)

        if match:
            if current_msg["datetime"] is not None:
                messages.append(current_msg)
                current_msg = {
                    "datetime": None,
                    "date": None,
                    "time": None,
                    "sender": None,
                    "message": "",
                    "is_system": False
                }

            date_str = match.group(1)
            time_str = match.group(2)
            am_pm   = match.group(3)
            rest    = match.group(4)

            if am_pm:
                dt_str = f"{date_str} {time_str} {am_pm}"
                dt_formats = ["%d/%m/%y %I:%M %p", "%d/%m/%Y %I:%M %p"]
            else:
                dt_str = f"{date_str} {time_str}"
                dt_formats = ["%d/%m/%y %H:%M", "%d/%m/%Y %H:%M"]

            dt_obj = None
            for fmt in dt_formats:
                try:
                    dt_obj = datetime.strptime(dt_str, fmt)
                    break
                except:
                    pass

            current_msg["datetime"] = dt_obj
            current_msg["date"] = dt_obj.date() if dt_obj else None
            current_msg["time"] = dt_obj.time() if dt_obj else None

            if ": " in rest:
                sender, msg = rest.split(": ", 1)
                current_msg["sender"] = sender.strip()
                current_msg["message"] = msg.strip()
                current_msg["is_system"] = False
            else:
                current_msg["sender"] = None
                current_msg["message"] = rest.strip()
                current_msg["is_system"] = True

        else:
            if current_msg["message"]:
                current_msg["message"] += " " + line
            else:
                current_msg["message"] = line

    if current_msg["datetime"] is not None:
        messages.append(current_msg)

    df = pd.DataFrame(messages)
    return df


# =======================================================
# LOAD WHATSAPP CHAT FILE
# =======================================================

file_path = "data/chat.txt"   # change if needed

with open(file_path, "r", encoding="utf-8") as f:
    raw_lines = f.readlines()

print("‚úÖ Total Raw Lines:", len(raw_lines))


# =======================================================
# PARSE CHAT INTO df
# =======================================================

df = parse_whatsapp_chat(raw_lines)
print("‚úÖ Parsed Messages:", len(df))


# =======================================================
# FINAL CLEANING (df REMAINS THE SAME NAME ‚úÖ)
# =======================================================

# Remove system messages
df = df[df["is_system"] == False]

# Remove None senders
df = df[df["sender"].notna()]

# Remove empty messages
df = df[df["message"].notna()]
df = df[df["message"].str.strip() != ""]

# Reset index
df.reset_index(drop=True, inplace=True)


# =======================================================
# FINAL VERIFICATION
# =======================================================

print("\n‚úÖ FINAL CLEAN DATA SHAPE:", df.shape)
print("\n‚úÖ NULL CHECK:\n")
print(df.isnull().sum())

df.head()
```

```python
# ============================================================
# WHATAI ‚Äì WHATSAPP CHAT NLP INTELLIGENCE SYSTEM (ONE FILE)
# BACKEND / IPYTHON VERSION
# ============================================================

# ---- INSTALLS (UNCOMMENT AND RUN ONCE IF NEEDED) ----------
# !pip install nltk
# !pip install wordcloud
# !pip install nrclex

import re
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
from wordcloud import WordCloud
from nrclex import NRCLex

# ------------------------------------------------------------
# 1. LOAD RAW WHATSAPP CHAT (ANDROID EXPORT)
# ------------------------------------------------------------

file_path = "data/chat.txt"   # üîÅ change if needed

with open(file_path, "r", encoding="utf-8") as f:
    raw_lines = f.readlines()

print("‚úÖ Total raw lines:", len(raw_lines))


# ------------------------------------------------------------
# 2. PARSE ANDROID WHATSAPP CHAT
# ------------------------------------------------------------

# Example lines:
# 12/10/24, 9:15 pm - Name: Message
# 12/10/2024, 21:15 - Name: Message

msg_pattern = re.compile(
    r'^(\d{1,2}/\d{1,2}/\d{2,4}),\s(\d{1,2}:\d{2})\s?(am|pm|AM|PM)?\s-\s(.*)$'
)

def parse_whatsapp_chat(lines):
    messages = []

    current_msg = {
        "datetime": None,
        "date": None,
        "time": None,
        "sender": None,
        "message": "",
        "is_system": False
    }

    for raw_line in lines:
        line = raw_line.strip()

        match = msg_pattern.match(line)

        if match:
            # save previous message
            if current_msg["datetime"] is not None:
                messages.append(current_msg)
                current_msg = {
                    "datetime": None,
                    "date": None,
                    "time": None,
                    "sender": None,
                    "message": "",
                    "is_system": False
                }

            date_str = match.group(1)
            time_str = match.group(2)
            am_pm   = match.group(3)
            rest    = match.group(4)

            # datetime parsing
            if am_pm:
                dt_str = f"{date_str} {time_str} {am_pm}"
                dt_formats = ["%d/%m/%y %I:%M %p", "%d/%m/%Y %I:%M %p"]
            else:
                dt_str = f"{date_str} {time_str}"
                dt_formats = ["%d/%m/%y %H:%M", "%d/%m/%Y %H:%M"]

            dt_obj = None
            for fmt in dt_formats:
                try:
                    dt_obj = datetime.strptime(dt_str, fmt)
                    break
                except:
                    pass

            current_msg["datetime"] = dt_obj
            current_msg["date"] = dt_obj.date() if dt_obj else None
            current_msg["time"] = dt_obj.time() if dt_obj else None

            # sender + message
            if ": " in rest:
                sender, msg = rest.split(": ", 1)
                current_msg["sender"] = sender.strip()
                current_msg["message"] = msg.strip()
                current_msg["is_system"] = False
            else:
                current_msg["sender"] = None
                current_msg["message"] = rest.strip()
                current_msg["is_system"] = True

        else:
            # multi-line continuation
            if current_msg["message"]:
                current_msg["message"] += " " + line
            else:
                current_msg["message"] = line

    # append last message
    if current_msg["datetime"] is not None:
        messages.append(current_msg)

    df_parsed = pd.DataFrame(messages)
    return df_parsed

df = parse_whatsapp_chat(raw_lines)
print("‚úÖ Parsed messages:", len(df))
print(df.head())


# ------------------------------------------------------------
# 3. CLEAN DATA (REMOVE SYSTEM / EMPTY / NONE)
# ------------------------------------------------------------

# remove system messages
df = df[df["is_system"] == False]

# remove None senders
df = df[df["sender"].notna()]

# remove empty messages
df = df[df["message"].notna()]
df = df[df["message"].str.strip() != ""]

df.reset_index(drop=True, inplace=True)

print("\n‚úÖ After cleaning, messages:", len(df))
print(df.head())


# ------------------------------------------------------------
# 4. SENTIMENT ANALYSIS (VADER)
# ------------------------------------------------------------

nltk.download("vader_lexicon")
sid = SentimentIntensityAnalyzer()

df["sentiment_scores"] = df["message"].apply(lambda x: sid.polarity_scores(str(x)))

df["neg"] = df["sentiment_scores"].apply(lambda x: x["neg"])
df["neu"] = df["sentiment_scores"].apply(lambda x: x["neu"])
df["pos"] = df["sentiment_scores"].apply(lambda x: x["pos"])
df["compound"] = df["sentiment_scores"].apply(lambda x: x["compound"])

def label_sentiment(score):
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df["sentiment_label"] = df["compound"].apply(label_sentiment)

print("\n‚úÖ Sentiment columns added:")
print(df[["sender", "message", "neg", "neu", "pos", "compound", "sentiment_label"]].head())

# simple sentiment bar chart
sentiment_counts = df["sentiment_label"].value_counts()
plt.figure()
sentiment_counts.plot(kind="bar")
plt.title("Overall Chat Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Number of Messages")
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# 5. TOPIC MODELING (TF-IDF + NMF + WORDCLOUD)
# ------------------------------------------------------------

tfidf = TfidfVectorizer(
    max_df=0.95,
    min_df=2,
    stop_words="english"
)

dtm = tfidf.fit_transform(df["message"])

num_topics = 5
nmf_model = NMF(n_components=num_topics, random_state=42)
nmf_model.fit(dtm)

feature_names = tfidf.get_feature_names_out()

for index, topic in enumerate(nmf_model.components_):
    print(f"\nüü¢ TOPIC #{index + 1}")

    top_words_idx = topic.argsort()[-15:]
    topic_words = {feature_names[i]: topic[i] for i in top_words_idx}

    print("Top words:", list(topic_words.keys()))

    wc = WordCloud(
        width=800,
        height=400,
        background_color="white"
    ).generate_from_frequencies(topic_words)

    plt.figure(figsize=(10, 5))
    plt.imshow(wc)
    plt.axis("off")
    plt.title(f"Topic {index + 1}")
    plt.show()

# assign dominant topic to each message
topic_values = nmf_model.transform(dtm)
df["dominant_topic"] = topic_values.argmax(axis=1)

# topic-wise average sentiment
topic_sentiment = df.groupby("dominant_topic")["compound"].mean().reset_index()
topic_sentiment.columns = ["Topic", "Average_Sentiment"]

print("\n‚úÖ Topic-wise average sentiment:")
print(topic_sentiment)

plt.figure()
topic_sentiment.set_index("Topic")["Average_Sentiment"].plot(kind="bar")
plt.title("Topic-wise Average Sentiment (Compound Score)")
plt.xlabel("Topic")
plt.ylabel("Average Sentiment")
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# 6. EMOTION DETECTION (NRCLex)
# ------------------------------------------------------------

def detect_emotion(text):
    try:
        emotion_scores = NRCLex(str(text)).raw_emotion_scores
        if emotion_scores:
            return max(emotion_scores, key=emotion_scores.get)
        else:
            return "neutral"
    except:
        return "neutral"

df["emotion"] = df["message"].apply(detect_emotion)

print("\n‚úÖ Emotion column added:")
print(df[["sender", "message", "emotion"]].head())

emotion_counts = df["emotion"].value_counts()
plt.figure()
emotion_counts.plot(kind="bar")
plt.title("Emotion Distribution in WhatsApp Chat")
plt.xlabel("Emotion")
plt.ylabel("Number of Messages")
plt.tight_layout()
plt.show()


# ------------------------------------------------------------
# 7. ADVANCED FILTERS + AI INSIGHT GENERATOR
# ------------------------------------------------------------

def filter_by_user(df_in, user):
    if user is None or user == "All":
        return df_in
    return df_in[df_in["sender"] == user]

def filter_by_date_range(df_in, start_date=None, end_date=None):
    temp = df_in.copy()
    if start_date:
        temp = temp[temp["date"] >= pd.to_datetime(start_date).date()]
    if end_date:
        temp = temp[temp["date"] <= pd.to_datetime(end_date).date()]
    return temp

def filter_by_keyword(df_in, keyword=None):
    if keyword is None or keyword.strip() == "":
        return df_in
    return df_in[df_in["message"].str.contains(keyword, case=False, na=False)]

def generate_ai_insights(df_in):
    insights = []
    if df_in.empty:
        return ["No messages available for insights."]

    # Most active user
    try:
        top_user = df_in["sender"].value_counts().idxmax()
        insights.append(f"üë§ Most active user: {top_user}")
    except:
        pass

    # Overall sentiment
    try:
        sentiment_summary = df_in["sentiment_label"].value_counts().idxmax()
        insights.append(f"üòä Overall chat mood: {sentiment_summary}")
    except:
        pass

    # Most discussed topic
    if "dominant_topic" in df_in.columns:
        try:
            top_topic = df_in["dominant_topic"].value_counts().idxmax()
            insights.append(f"üß† Most discussed topic: Topic {top_topic}")
        except:
            pass

    # Most common emotion
    try:
        top_emotion = df_in["emotion"].value_counts().idxmax()
        insights.append(f"‚ù§Ô∏è Most common emotion: {top_emotion}")
    except:
        pass

    # Peak chat hour
    try:
        # convert time to string hour "HH"
        hour_series = df_in["time"].astype(str).str[:2]
        peak_hour = hour_series.value_counts().idxmax()
        insights.append(f"‚è∞ Peak chatting hour: {peak_hour}:00")
    except:
        pass

    return insights

# Example usage on full data
df_filtered = filter_by_user(df, "All")
df_filtered = filter_by_keyword(df_filtered, None)
df_filtered = filter_by_date_range(df_filtered, None, None)

ai_insights = generate_ai_insights(df_filtered)

print("\n‚úÖ AI AUTO INSIGHTS:\n")
for ins in ai_insights:
    print("‚Ä¢", ins)


# ------------------------------------------------------------
# 8. SAVE FINAL DATAFRAME FOR STREAMLIT / FURTHER USE
# ------------------------------------------------------------

df.to_csv("whatsapp_final.csv", index=False)
print("\n‚úÖ Final data saved to whatsapp_final.csv")

```
