```python
#Setup Gemini
import google.generativeai as genai

genai.configure(api_key="google gemini key")

MODEL_NAME = "models/gemini-flash-latest"
```

```python
#Create Agent class
class GeminiAgent:
    def __init__(self, role_name, system_prompt):
        self.role_name = role_name
        self.model = genai.GenerativeModel(
            MODEL_NAME,
            system_instruction=system_prompt
        )

    def run(self, message):
        response = self.model.generate_content(message)
        return response.text
```

```python
#Agent 1: Planner
planner_agent = GeminiAgent(
    role_name="Planner",
    system_prompt="""
You are a planning agent.
Your job is to break the problem into clear logical steps.
Do NOT write code.
"""
)

```

```python
#Agent 2: Code Write
code_agent = GeminiAgent(
    role_name="CodeWriter",
    system_prompt="""
You are a Python coding agent.
Write clean, optimized Python code.
Print final results.
"""
)
```

```python
#Agent-to-Agent flow
# User task
task = "Calculate permutations of the word ALGEbRA and give final count."

# Step 1: Planner thinks
plan = planner_agent.run(task)

print("PLANNER OUTPUT:")
print(plan)
print("-" * 50)

# Step 2: Code agent executes plan
code_result = code_agent.run(
    f"Based on this plan, write Python code and solve the task:\n\n{plan}"
)

print("CODE AGENT OUTPUT:")
print(code_result)
```
